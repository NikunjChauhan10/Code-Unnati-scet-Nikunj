```python
x=[2,4,6,8,10]
y=[1,2,-3,-4,-5]
```

```python
#X label
plt.xlabel("sales")

#y label
plt.ylabel("profit")

#plot values 
plt.plot(x,y)

#plot title
plt.title("sales vs Profit")
```

```python
#X label
plt.xlabel("sales")

#y label
plt.ylabel("profit")

#plot values 
plt.plot(x,y,linestyle='dotted')

#plot title
plt.title("sales vs Profit")
```

```python
#X label
plt.xlabel("sales")

#y label
plt.ylabel("profit")

#plot values 
plt.plot(x,y,color="r")

#plot title
plt.title("sales vs Profit")
```

```python
#X label
plt.xlabel("sales")

#y label
plt.ylabel("profit")

#plot values 
plt.plot(x,y,color="r",linestyle='dotted',marker='o')

#plot title
plt.title("sales vs Profit")
```

```python
#X label
plt.xlabel("sales")

#y label
plt.ylabel("profit")

#plot values 
plt.plot(x,y,color="b",marker='+',ms=15)

#plot title
plt.title("sales vs Profit")
```

```python
#X label
plt.xlabel("sales")

#y label
plt.ylabel("profit")

#plot title
plt.title("sales vs Profit")

#plot values
plt.bar(x,y)
```

```python
#X label
plt.xlabel("sales")

#y label
plt.ylabel("profit")

#plot title
plt.title("sales vs Profit")

#plot values
plt.scatter(x,y,marker= '+')
```
